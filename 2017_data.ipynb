# Creating data of 2017 with similar columns as data upto 2016

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', None)
```

```python
credits = pd.read_csv('credits.csv')

credits
```

```python
meta = pd.read_csv('movies_metadata.csv')
meta
```

```python
meta.dtypes
```

```python
meta['release_date'] =pd.to_datetime(meta['release_date'] , errors='coerce')
meta
```

```python
meta['year'] = meta['release_date'].dt.year
# meta['month'] = meta['release_date'].dt.month
meta['year'].value_counts().sort_index()
```

```python
new_meta = meta.loc[meta.year ==2017 , ['genres' , 'id','title' , 'year']]
new_meta
```

```python
new_meta['id'] = new_meta['id'].astype(int)
new_meta.dtypes
```

```python
data = pd.merge(new_meta , credits , on='id')
data
```

```python
import ast
data['genres'] = data['genres'].map(lambda x : ast.literal_eval(x))
data['cast'] = data['cast'].map(lambda x : ast.literal_eval(x))
data['crew'] = data['crew'].map(lambda x : ast.literal_eval(x))
data
```

```python
def make_gernesList(x):
    gen = []
    st = ' '
    for i in x:
        if i.get('name')=='Science Fiction':
            scifi = 'Sci-Fi'
            gen.append(scifi)
        else:
            gen.append(i.get('name'))
    if gen==[]:
        return np.NaN
    else:
        return st.join(gen)
```

```python
data['genres_list'] = data['genres'].map(lambda x: make_gernesList(x))
data
```

```python
data['cast'][0]
```

```python
def get_actor1(x):
    cast =[]
    for i in x:
        cast.append(i.get('name'))
    if cast ==[]:
        return np.NaN
    else:
        return cast[0]
```

```python
data['actor_1_name'] = data['cast'].map(lambda x: get_actor1(x))
data
```

```python
def get_actor2(x):
    cast =[]
    for i in x:
        cast.append(i.get('name'))
    if cast ==[] or len(cast) <=1 :
        return np.NaN
    else:
        return cast[1]
```

```python
data['actor_2_name'] = data['cast'].map(lambda x: get_actor2(x))
data
```

```python
def get_actor3(x):
    cast =[]
    for i in x:
        cast.append(i.get('name'))
    if cast ==[] or len(cast) <=2 :
        return np.NaN
    else:
        return cast[2]
```

```python
data['actor_3_name'] = data['cast'].map(lambda x: get_actor3(x))
data
```

```python
def get_directors(x):
    dt = []
    st = " "
    for i in x:
        if i.get('job') == 'Director':
            dt.append(i.get('name'))
    if dt == []:
        return np.NaN
    else:
        return (st.join(dt))
```

```python
data['director_name'] = data['crew'].map(lambda x: get_directors(x))
data
```

```python
movie = data.loc[: , ['director_name','actor_1_name','actor_2_name','actor_3_name','genres_list','title']]
movie
```

```python
movie.isna().sum()
```

```python
movie.shape
```

```python
movie = movie.dropna(how = 'any')
```

```python
movie.isna().sum()
```

```python
movie.shape
```

```python
movie = movie.rename(columns={'genres_list':'genres' , 'title':'movie_title'})
movie['movie_title'] = movie['movie_title'].str.lower()
movie['comb'] = movie['actor_1_name'] + ' ' + movie['actor_2_name'] + ' '+ movie['actor_3_name'] + ' '+ movie['director_name'] +' ' + movie['genres']
movie
```

```python
old = pd.read_csv('data1.csv')
old
```

```python
old['comb'] = old['actor_1_name'] + ' ' + old['actor_2_name'] + ' '+ old['actor_3_name'] + ' '+ old['director_name'] +' ' + old['genres']
old
```

```python
new = pd.concat([old , movie] , axis=0)
new.shape
```

```python
new.drop_duplicates(subset='movie_title' , keep='last', inplace=True)
new.shape
```

```python
new.to_csv('new_data' , index=False)
```
